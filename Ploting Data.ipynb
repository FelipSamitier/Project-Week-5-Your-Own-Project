```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
%matplotlib inline
```

```python
# Human Rights by Year

spain = pd.read_csv("final_data/Years_Human_Rights_Spain.csv")
norway = pd.read_csv("final_data/Years_Human_Rights_Norway.csv")
czechia = pd.read_csv("final_data/Years_Human_Rights_Czechia.csv")

fig,ax = plt.subplots(figsize = (7,7))

ax.set_title("Human Rights", fontsize = 20)

ax.plot(spain["Year"],spain["Human Rights"],label = "Spain", color = "darkgoldenrod")
ax.plot(spain["Year"],norway["Human Rights"],label = "Norway", color = "midnightblue")
ax.plot(spain["Year"],czechia["Human Rights"],label = "Czechia",color = "firebrick")

ax.legend()
```

```python
# Corruption Index of Each Country by Year

spain2 = pd.read_csv("final_data/year_schooling_vs_corruption_Spain.csv")
norway2 = pd.read_csv("final_data/year_schooling_vs_corruption_Norway.csv")
czechia2 = pd.read_csv("final_data/year_schooling_vs_corruption_Czechia.csv")

spain3 = spain2.drop(["Unnamed: 0"], axis = 1)
norway3 = norway2.drop(["Unnamed: 0"], axis = 1)
czechia3 = czechia2.drop(["Unnamed: 0"], axis = 1)

fig,ax = plt.subplots(figsize = (7,7))

ax.set_title("Perception People of Government Corruption", fontsize = 20)

ax.plot(spain3["Year"],100 - spain3["Corruption_Perception"],label = "Spain", color = "darkgoldenrod")
ax.plot(spain3["Year"],100 - norway3["Corruption_Perception"],label = "Norway", color = "midnightblue")
ax.plot(spain3["Year"],100 - czechia3["Corruption_Perception"],label = "Czechia", color = "firebrick")

ax.legend()
```
