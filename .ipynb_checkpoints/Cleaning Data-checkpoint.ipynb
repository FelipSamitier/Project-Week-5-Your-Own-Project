```python
# Relación entre education y democracia (Czechia)

data9 = pd.read_csv("data/Stats Arregladas/correlation-between-education-and-democracy.csv")

data9_prova = data9[data9["Entity"] == "Czechia"]

data9_plot = data9_prova[data9_prova["Year"] >= 1940]

data9_plot = data9_plot.drop(["Total population (Gapminder, HYDE & UN)","Continent","Code","Entity"],axis=1)
data9_prova2 = data9_plot.rename(columns={"Total years of schooling (Lee-Lee (2016))" : "Total_Years_Education",
                          "Political Regime (OWID based on Polity IV and Wimmer & Min)":"Political_Regime"})


data9_prova2.drop(data9_prova2[data9_prova2.Total_Years_Education.isnull()].index, inplace = True)

data9_prova2.to_csv("year_PoliticalRegime_YearsEducation_Czechia.csv")
```

```python
# Derechos Humanos (Spain)

data10 = pd.read_csv("data/Stats Arregladas/human-rights-scores.csv")

data10 = data10.rename(columns={"Human Rights Score (Schnakenberg & Fariss, 2014; Fariss, 2019)" : "Human Rights"})

data10_prova = data10[data10["Entity"] == "Spain"]

data10_plot = data10_prova.drop(["Code","Entity"],axis = 1)

fig,ax = plt.subplots(figsize = (7,7))

ax.set_title("Prova Plot 1 - Puntuació Drets Humans, Espanya 1946 al 2017", fontsize = 20)

ax.plot(data10_plot["Year"],data10_plot["Human Rights"],label = "Puntuacio drets humans")

ax.legend()

data10_plot.to_csv("Years_Human_Rights_Spain.csv")
```

```python
# Years of schooling (Czehia)

Years_of_schooling = pd.read_csv("data/Stats Arregladas/mean-years-of-schooling-1.csv")

Years_of_schooling_1 = Years_of_schooling[Years_of_schooling["Entity"] == "Czechia"]

Years_of_schooling_2 = Years_of_schooling_1[Years_of_schooling_1["Year"] >= 1940]

Years_of_schooling_3 = Years_of_schooling_2.drop(["Code"],axis=1)
Years_of_schooling_4 = Years_of_schooling_3.rename(columns={
    "Average Total Years of Schooling for Adult Population (Lee-Lee (2016), Barro-Lee (2018) and UNDP (2018))": 
    "Average_Years_Education"})

Years_of_schooling_4.to_csv("year_schooling_Czechia.csv")
```

```python
share_of_education = pd.read_csv("data/Stats Arregladas/share-of-education-in-government-expenditure.csv")

share_of_education_1 = share_of_education[(share_of_education["Entity"] == "Czechia") | 
                                                (share_of_education["Entity"] == "Spain") | 
                                                (share_of_education["Entity"] == "Norway")]

share_of_education_1.dropna(1)

share_of_education_2 = share_of_education_1.rename(columns={
    "Share of education in government expenditure (% of government expenditure) (World Bank (2019))" : 
    "Share of education"
})

share_of_education_3 = share_of_education_2.drop(["Code"],axis = 1)

share_of_education_3_spain = share_of_education_3[(share_of_education_3["Entity"] == "Spain")]
share_of_education_3_norway = share_of_education_3[(share_of_education_3["Entity"] == "Norway")]
share_of_education_3_czechia = share_of_education_3[(share_of_education_3["Entity"] == "Czechia")]

share_of_education_3_spain.to_csv("year_share_education_spain.csv")
share_of_education_3_norway.to_csv("year_share_education_norway.csv")
share_of_education_3_czechia.to_csv("year_share_education_czechia.csv")
```
